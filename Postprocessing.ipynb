```python
#Relationship object

class Relationship_object:
    def __init__(self, sentence_id, source_word, source_word_position, target_words, target_words_positions):
        #Takes in the source text word and the target word or words it is associated with. 
        #Stores each as a list of strings.
        
        self.sentence_id = sentence_id
        self.source_word = source_word
        self.target_words = target_words
        self.source_word_position = source_word_position
        self.target_words_positions = target_words_positions
        
        
    def print(self):
        print("Source word: " + self.source_word)
        print("Source word position: " + str(self.source_word_position))
        p = "Target words: "
        for word in self.target_words:
            p = p + word + " "
        
        print(p)
        
        p2 = "Target words' positions: "
        for position in self.target_words_positions:
            p2 = p2 + str(position) + " "
        
        print(p2+ '\n')
        

```

```python
#Sentence pair object

class Sentence_pair:
    def __init__(self, text_representation, source_language, target_language):
        
        #Takes in the text representation of the alignment from the .A3 file and splits it into the component parts
        #****REQUIRES SPLITTING THE .A3 FILE ON "#" FIRST***
        
        #set defaults for attributes
        self.id = -1
        self.source_language = source_language
        self.target_language = target_language
        self.source_length = -1
        self.target_length = -1
        self.source_sentence = ("Default", "source", "sentence")
        self.target_sentence = ("Default", "target", "sentence")
        self. alignment = []
        self.alignment_score = -1
        
        #break the text representation at returns
        text = text_representation.split("\n")
        #print(text)
        
        #store target text w/o trailing whitespace
        self.target_sentence = text[1].strip().split()
        
        #break up first string
        info = text[0].split()
        #print(info)
        self.id = info[2][1:-1]
        self.source_length = info[5]
        self.target_length = info[8]
        self.alignment_score = info[-1]
        
        #store and strip out relationships from the source sentence
        source = text[2].strip().split(')')
        source_clean = []
        relationship_positions = []
        count = 0
        for word in source:
            temp = word.split('(')
            source_clean.append(temp[0].strip())
            relationship_positions.append(temp[-1].strip())
            
            
        #store source text
        #print(source_clean)
        self.source_sentence = source_clean[1:-1] #removes null and extra blank string
        
        
        #create relationship objects
        
        #set defaults
        sentence_id = self.id
        source_word = "default"
        target_words = []
        source_word_position = -1
        target_words_positions = [-1]
        
        relationships = []
        i = 0
        for word in self.source_sentence:
            source_word = word
            source_word_position = i
            target_words_positions = relationship_positions[i+1].split()
            #print(target_words_positions)
            for position in target_words_positions:
                if position.isnumeric():
                    target_words.append(self.target_sentence[int(position)-1])
                    #print(target_words)
            
            r = Relationship_object(sentence_id, source_word, source_word_position, target_words, target_words_positions)
            relationships.append(r)
            i = i + 1
            target_words = []
            
            
        self.alignment = relationships
        
    def get_source_sentence(self):
        temp = ''
        for word in self.source_sentence:
            temp = temp + word + ' '
            
        return temp.strip()
    
    
    def get_target_sentence(self):
        temp = ''
        for word in self.target_sentence: 
            temp = temp + word + ' '
            
        return temp.strip()
    
    
    def info(self):
        print('Sentence ID: ' + str(self.id))
        print('Source sentence: ' + str(self.source_sentence))
        print('Target sentence: ' + str(self.target_sentence))
        print('Source language: ' + str(self.source_language))
        print('Target language: ' + str(self.target_language))
        print('Source length: ' + str(self.source_length))
        print('Target length: ' + str(self.target_length))
        print('Alignment score: ' + str(self.alignment_score)+'\n')
        print('Alignment\n---------')
        for relationship in self.alignment: 
            relationship.print()    
        
        
```

```python
#output

#format into XML(ish) 

def output_XML(alignments, filename):
    cleanfile = open('testing/'+filename+'.xml', 'w', encoding="utf-8") 
    cleanfile.write("""<text>\n\t<body>\n\t\t<div xml:lang="eng and grc" type="alignment" n="URN HERE">\n""")
    for item in alignments:

        cleanfile.write('''\t\t\t<div type="textpart" n="'''  + item.id+'''" subtype="sentence_pair">\n''')
        cleanfile.write('''\t\t\t\t<p type="sourcesentence">'''+item.get_source_sentence()+'''</p>\n''')
        cleanfile.write('''\t\t\t\t<p type="targetsentence">'''+item.get_target_sentence()+'''</p>\n''')
        cleanfile.write('''\t\t\t\t<p type="sourcelength">'''+str(item.source_length)+'''</p>\n''')
        cleanfile.write('''\t\t\t\t<p type="targetlength">'''+str(item.target_length)+'''</p>\n''')
        cleanfile.write('''\t\t\t\t<p type="alignmentscore">'''+str(item.alignment_score)+'''</p>\n''')
        
        cleanfile.write('''\t\t\t</div>\n''')
        
    cleanfile.write("""\t\t</div>\n\t</body>\n</text>""")
    
    cleanfile.close()
        
    print('Success!')
    
    
#JSON(?)

#English [greek] english [greek](?)
```

```python
for work in listOfFiles:

    file = open(os.path.join(path, work), "r", encoding="utf-8")
    content = file.read()
    file.close()
    
    #split on '#'
    
    alignments_raw = content.split('#')
    alignments_raw = alignments_raw[1:]
    #print(alignments_raw[0])

    #determine direction of translation, set source and target languages
    if 'reverse' in work:
        source = 'Greek'
        target = 'English'
    else:
        source = 'English'
        target = 'Greek'

    #make Sentence_pair for each
    alignments = []
    for text in alignments_raw: 
        processed_text = Sentence_pair(text, source, target)
        alignments.append(processed_text)
    
    #store as various file types
    output_XML(alignments, work[:-9]+'.test1.xml')
    
    
    #testing
    #alignments[0].info() 
    #alignments[-1].info()
```
